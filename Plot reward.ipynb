```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib 
```

```python
def f(x,y):
    # alias
    x2 = x
    y2 = y
    # enemy
    x4 = 350
    y4 = 350

    f2 = 0 

    Ax4, Ay4 = -np.cos(f2), np.sin(f2)
    Bx24, By24 = x2 - x4, y2 - y4

    dist = - np.sqrt(np.abs((x2-x4)**2 + (y2-y4)**2))

    phy = (Ax4*Bx24 + Ay4*By24)/(np.sqrt(Ax4**2 + Ay4**2) * np.sqrt(Bx24**2 + By24**2))
    
    out = phy*(dist+500)#+300

    return out
```

```python
xline = np.linspace(0, 500,100)
yline = np.linspace(0, 500,100)
X, Y = np.meshgrid(xline, yline)
Z = f(X, Y)
# Z
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
